#### 9.Find the month that has the highest sales and the highest amount spend?

```python


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Convert 'Date' column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])

# Extract month from 'Date' column
df1['Month'] = df1['Date'].dt.month

# Group data by 'Month' and sum 'Sales'
monthly_sales = df1.groupby('Month', as_index=False)['Quantity'].sum()

# Sort and find the month with the highest sales
highest_sales_month = monthly_sales.sort_values(by='Quantity', ascending=False).head(1)

# Seaborn bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Quantity', data=monthly_sales)

# Highlight highest sales month
sns.barplot(x=highest_sales_month['Month'], y=highest_sales_month['Quantity'], color='red')

# Annotate highest sales
plt.annotate(f'Highest Sales: {highest_sales_month["Quantity"].values[0]}', 
             xy=(highest_sales_month['Month'].values[0], highest_sales_month['Quantity'].values[0]), 
             textcoords='offset points', xytext=(0,10), ha='center')

plt.xlabel('Month')
plt.ylabel('Sales Quantity')
plt.title('Monthly Sales')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

August has the highest sales with the total amount of 27379.0

#### 13.How many male customers brought the Nest Learning Thermostat 3rd Gen-USA - Stainless Steel

```python


import seaborn as sns
import matplotlib.pyplot as plt

# Filter data for Nest thermostat sales
nest_thermostat_sales = df[df['Product_Description'] == 'Nest Learning Thermostat 3rd Gen-USA - Stainless Steel']

# Group by Gender and count
count_males = nest_thermostat_sales.groupby('Gender', as_index=False)['Product_Description'].count()
count_males = count_males.rename(columns={'Product_Description': 'Count'})
print(count_males)
# Seaborn pie chart
plt.figure(figsize=(6, 6))
sns.set_style("whitegrid")
plt.pie(count_males['Count'], labels = count_males['Gender'], autopct='%1.1f%%', 
        wedgeprops={'edgecolor':'black'}, 
        textprops={'fontsize':12})
plt.title('Nest Learning Thermostat 3rd Gen-USA - Stainless Steel Sales by Gender')
plt.legend(labels=count_males['Gender'], 
           title="Gender", 
           bbox_to_anchor=(1,0.5), 
           loc="center right", 
           bbox_transform=plt.gcf().transFigure)
plt.tight_layout()
plt.show()
```

1375 male customer brought Nest Learning Thermostat 3rd Gen-USA - Stainless Steel with a percentage of 39.2%

14.Which coupon code  has the highest delivery charges? 

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Group by Coupon Code, calculate mean Delivery Charges, and get top 5 coupons
top_5_coupon = df1.groupby('Coupon_Code', as_index=False)['Delivery_Charges'].mean().reset_index().nlargest(5, 'Delivery_Charges')
print(top_5_coupon)
# Plot bar chart
plt.figure(figsize=(10, 6))

# Use seaborn's color palette
sns.barplot(x='Coupon_Code', y='Delivery_Charges', data=top_5_coupon, 
            palette=sns.color_palette("husl", len(top_5_coupon)))

plt.title('Top 5 Coupons by Average Delivery Charges')
plt.xlabel('Coupon Code')
plt.ylabel('Average Delivery Charges')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

15.Which gender has used coupon mostly

```python


import matplotlib.pyplot as plt

# Filter data for used coupons
most_used = df1[df1['Coupon_Status'] == 'Used'].groupby('Gender', as_index=False)['Coupon_Status'].count().reset_index()

# Rename column for clarity
most_used = most_used.rename(columns={'Coupon_Status': 'Number_of_People_Who_Used_Coupons'})

# Print result
print(most_used)

# Plot donut chart
plt.figure(figsize=(8, 8))
plt.pie(most_used['Number_of_People_Who_Used_Coupons'], labels = most_used['Gender'], autopct='%1.1f%%', 
        pctdistance=0.85, startangle=90)

# Create a white circle in the middle to make it look like a donut
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')  

plt.title('Coupon Usage by Gender')
plt.show()
```

Females have used the coupon mostly with the number of 6705
